# 데이터의 종류
> 데이터를 시계열자료, 횡단면자료, 패널자료로 구분한다.

## 시계열자료
---

시간의 변화에 따라 얻는 자료를 **시계열자료(time-series data)** 라고 한다.

예를 들어 연도별 이자율, 월별 평균 온도, 매 4년마다 우리나라 올림픽 메달 수는 시계열자료이다.

## 횡단면자료
---

동시간 대에 얻는 자료를 **횡단면자료(cross-sectional data)** 라고 한다.

예를 들어 은행별 이자율, 도시별 온도, 역대 올림픽에서 금메달을 획득한 종목은 횡단면자료이다.

## 패널자료
---

횡단면자료가 시간에 따라 나열되어 있는 것이 **패널자료(panel data)** 이다.

예를 들어 아파트 가격이 지역별, 연도별로 되어 있으면 패널자료이다.

# R에서 시계열 그래프 그리기
> R 함수와 그래프 그리는 방법 소개

```r
stocks <- scan("istat_data\\SamsungElectronicsStocks.txt")
times  <- c(1:length(stocks))
```

```r
print(head(times))
```

**head** 함수는 처음 6개의 데이터를 출력하는 함수이다.

```r
print(seq(1, 10, 3))
```

```r
print(seq(from = 1, to = 10, by = 3))
```

**seq(a, b, c)** 또는 **seq(from = a, to = b, by = c)** 를 사용하면 숫자 **a** 부터 **b** 까지 간격이 **c**인 수열을 생성한다.

## **ts.plot** 함수를 이용한다.

```r
ts.plot(stocks, main = "삼성전자 주가")
```

## 일반 plot 함수에 옵션 **type = "l"** 을 추가한다.

```r
plot(times, stocks, main = "삼성전자 주가")
```

```r
par(mfcol = c(2,1))
plot(times, stocks)
```
