## Problem 4.2

```r
mean = c(14.3, -4.4, -6.1)
size = c(30, 29, 30)
sd = c(34.8, 31.9, 35.4)

g_mean = 1.33
g_sd = 35
g_size = sum(size)

SSG = sum(size*(mean - g_mean)**2)
SST = (g_size-1)*g_sd**2
SSE = SST - SSG

MSG = SSG / (length(size) - 1)
MSE = SSE / (g_size - length(size))

F = MSG / MSE

SSG
SST
SSE
MSG
MSE
F
```

## Problem 4.4

- 우선 SandwichAnts를 분산분석하자.

- 대략 대칭이고 이상점이 존재하지 않으므로 정규성을 만족한다.

- 가장 큰 표준편차가 가장 작은 표준편차의 2배 이하이므로 등분산 가정을 만족한다.

```r
mean = c(30.750, 34, 49.250)
sd = c(9.254, 14.629, 10.794)
size = c(8, 8, 8)

g_mean = 38
g_sd = 13.947
g_size = 24

SSG = sum(size*(mean - g_mean)**2)
SST = (g_size-1)*g_sd**2
SSE = SST - SSG

MSG = SSG / (length(size) - 1)
MSE = SSE / (g_size - length(size))

F = MSG/MSE

F
```

- F-통계량이 5.63임을 알았다.

- 땅콩버터와 햄피클의 평균 개미 수의 차이가 유의한지 5% 유의수준에서 검증하자

```r
t = (mean[2] - mean[3]) / sqrt(MSE*(1/size[2] + 1/size[3]))
t
```

- t-값이 -2.59임을 알았다.

- 자유도가 n-k(24-3=21)인 t-분포에서 p-값을 확인하자.

- 양쪽 꼬리 검증이므로 p-값은 0.0085*2 = 0.017이다.

- 유의수준 5%에서 영가설을 기각한다.
