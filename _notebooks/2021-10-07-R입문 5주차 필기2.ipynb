# ifelse()

## 기본 구조

```r
ifelse(조건, 조건이 TRUE이면 실행, 조건이 FALSE이면 실행)
```

## Problem

- 절댓값을 출력하는 코드를 작성하라

### Solution I
> 다소 복잡한 코드

- 아래 코드는 복잡하다는 생각이 든다.

````r
absolute <- function(x){
    for (i in 1:length(x)){
        if (x[i] > 0){
            x[i] = x[i]
        }else{
            x[i] = -x[i]
        }
    }
    print(x)
}

x = -10:10
plot(absolute(x))
````

### Solution II
> ifelse문을 사용한 간결한 코드

````r
friend <- function(x){
    ifelse(x>0, x, -x)
}

x = -10:10
friend(x)
````

## 이런 것도 가능

````r
y = ifelse(x < 0, -x, ifelse(x < 4, 0, x))

y
plot(y)
````

````r
plot(x,x)
points(x,y, col = 2, pch = 16)
````
